## Q. 2) Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ? 

```python
dataSpending = data[data['violation']== 'Speeding']
dataSpending
```

```python
index = dataSpending.groupby('driver_gender').violation.count().idxmax()
value = dataSpending.groupby('driver_gender').violation.count().max()
if index == 'M':
    indexv = 'Men'
else:
    indexv = 'Women'
```

```python
print(f"{indexv} are stopped more often with a value = {value} ")
```
